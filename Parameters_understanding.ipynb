## 2 - Where $R_{SN}$ is come from

To determine $\mathcal{R}_{SN}$, the nuclear AGN-corrected supernova rate per resolved galaxy
in the GOALS sample, we need a lot of data.  
First:

$$ \boxed{\frac{\mathcal{R}_{SN}}{\text{yr}^{-1}} = \Lambda _{IR} \left(\frac{\mathcal{G}\cdot[1-\langle\alpha_{AGN}\rangle]\cdot L_{IR}}{\text{erg}\cdot \text{s}^{-1}}\right)}$$

With $[1-\langle\alpha_{AGN}\rangle]$ the correction due to the IR Luminosity from AGN, $\mathcal{G}$ the fraction of IR luminosity produced by nuclear starburst activity and $\Lambda _{IR}$ the calibration factor.

|$\text{\underline{Parameters}}$|$\Lambda _{IR}$|$\mathcal{G}$|$[1-\langle\alpha_{AGN}\rangle]$|
|-|-|-|-|
|$\text{\underline{Def}}$|Calibration factor|Fraction of IR luminosity produced by nuclear starburst activity|Correction due to the IR Luminosity from AGN|
|$\text{\underline{Value}}$|$5.97 \times 10^{-46}$|0.5 used to not surestimate |~0.9|
|$\text{\underline{From}}$|Kroupa IMF|four GOals LIRGs|GOALs|

The calibration factor is define by two others $\Lambda _{IR}=A_{SFR}\cdot A_{IR}$ using the web-based
software Starburst99 (SB99).
